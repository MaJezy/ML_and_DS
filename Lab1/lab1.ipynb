```python
# Задание по строкам. Вариант 1
text = input("Введите русскоязычный текст: ")

# Приводим к нижнему регистру и разбиваем текст на слова
words = text.lower().split()

# Считаем слова, которые начинаются с "е"
count_e = sum(1 for word in words if word.startswith('е'))

print(f"Количество слов, начинающихся на букву 'е': {count_e}")
```

```python
# Массивы. Вариант 1. Задание 1
n = int(input("Введите длину целочисленного массива: "))
arr_int = []

for i in range(n):
    arr_int.append(int(input(f"Введите {i} элемент: ")))

# Поиск максимума и вывод в обратном порядке
max_val = max(arr_int)
print(f"Максимальный элемент: {max_val}")
print("Массив в обратном порядке:", arr_int[::-1])

print("-" * 30)

# Массивы. Вариант 1. Задание 2
# (Ввод вещественных чисел разделяется пробелом)
arr_float = [float(x) for x in input("Введите массив действительных чисел через пробел: ").split()]

# Расчет среднего арифметического
avg_val = sum(arr_float) / len(arr_float) if arr_float else 0

# Замена нулей на среднее значение
arr_float_mod = [avg_val if x == 0 else x for x in arr_float]

print(f"Среднее арифметическое: {avg_val}")
print("Преобразованный массив:", arr_float_mod)
```

```python
import math

# Функции. Вариант 1. Задание 1 (Площади фигур)
def area_circle(radius):
    return math.pi * (radius ** 2)

def area_rectangle(width, height):
    return width * height

def area_triangle(base, height):
    return 0.5 * base * height


# Функции. Вариант 1. Задание 2 (Обработка 3-х массивов)
def process_array(arr):
    arr_sum = sum(arr)
    arr_avg = arr_sum / len(arr) if arr else 0
    return arr_sum, arr_avg

# Пример работы с 3 массивами
arrays = [
    [1, 5, 8, 12],
    [3, 3, 3, 3, 3],
    [10, -2, 4]
]

for idx, arr in enumerate(arrays, 1):
    s, avg = process_array(arr)
    print(f"Массив {idx}: {arr} -> Сумма: {s}, Среднее: {avg:.2f}")
```

```python
# Матрицы. Вариант 1. Задание 1
# Создаем квадратную матрицу 3х3 для теста
matrix_A = [
    [5,  2, -3],
    [4,  8,  6],
    [-1, 0,  1]
]

sum_above = 0
count_above = 0
n_size = len(matrix_A)

for i in range(n_size):
    for j in range(i + 1, n_size): # Строго над главной диагональю
        if matrix_A[i][j] > 0:
            sum_above += matrix_A[i][j]
            count_above += 1

print(f"Над диагональю -> Сумма положительных: {sum_above}, Количество: {count_above}")
print("-" * 30)


# Матрицы. Вариант 1. Задание 2
matrix_B = [
    [10, 2, 5, -1, 4],
    [0, 9, 23, 11, 2],
    [7, 3,  1,  8, 6]
]

print("Исходная матрица B:")
for row in matrix_B: print(row)

for row in matrix_B:
    min_idx = row.index(min(row))
    max_idx = row.index(max(row))
    
    # Меняем местами минимум и максимум с крайними элементами строки
    # Важно: учитываем перекрестные индексы, если экстремумы стоят на краях
    
    # 1. Меняем минимальный с последним
    row[min_idx], row[-1] = row[-1], row[min_idx]
    
    # Корректируем индекс максимума, если он стоял на месте последнего элемента
    if max_idx == len(row) - 1:
        max_idx = min_idx
        
    # 2. Меняем максимальный с первым
    row[max_idx], row[0] = row[0], row[max_idx]

print("\nМодифицированная матрица B:")
for row in matrix_B: print(row)
```
