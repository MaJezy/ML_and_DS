```python
# 0. Импорты и загрузка данных
import pandas as pd
from scipy.stats import pearsonr
import re

# Используем PassengerId в качестве индекса
df = pd.read_csv('titanic/train.csv', index_col='PassengerId')

# Функция для корректного сохранения ответов без переноса строки на конце
def save_answer(filename, answer_str):
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(str(answer_str).strip())
    print(f"Ответ для {filename} успешно сохранен: '{answer_str}'")

```

```python
# 1. Количество мужчин и женщин
sex_counts = df['Sex'].value_counts()
# Ожидается формат: сначала мужчины, затем женщины через пробел
ans1 = f"{sex_counts['male']} {sex_counts['female']}"
save_answer('1.txt', ans1)
```

```python
# 2. Доля выживших пассажиров
# Находим среднее по колонке Survived (так как там 0 и 1) и переводим в проценты
survived_pct = df['Survived'].mean() * 100
ans2 = f"{survived_pct:.2f}"
save_answer('2.txt', ans2)
```

```python
# 3. Доля пассажиров первого класса
# Считаем отношение количества пассажиров 1-го класса к общему числу
pclass1_pct = (df['Pclass'] == 1).mean() * 100
ans3 = f"{pclass1_pct:.2f}"
save_answer('3.txt', ans3)
```

```python
# 4. Среднее и медиана возраста пассажиров
# Используются методы mean() и median() в pandas
age_mean = df['Age'].mean()
age_median = df['Age'].median()
ans4 = f"{age_mean:.2f} {age_median:.2f}"
save_answer('4.txt', ans4)
```

```python
# 5. Корреляция Пирсона между SibSp и Parch
# Считаем коэффициент корреляции Пирсона между числом братьев/сестер и родителей/детей
corr, _ = pearsonr(df['SibSp'], df['Parch'])
ans5 = f"{corr:.2f}"
save_answer('5.txt', ans5)
```

```python
# 6. Самое популярное женское имя
def get_female_first_name(full_name):
    # Убираем возможные пробелы по краям
    full_name = full_name.strip()
    
    # 1. Если это замужняя женщина "Mrs. ... (Имя ...)"
    # Ищем текст внутри скобок — там скрывается ее личное имя
    if "Mrs." in full_name and "(" in full_name:
        brackets_content = re.search(r'\((.*?)\)', full_name)
        if brackets_content:
            inner_name = brackets_content.group(1)
            # Убираем кавычки, если имя в духе: "Hannah" или ""Annie""
            inner_name = inner_name.replace('"', '').strip()
            # Берем первое слово внутри скобок (это и есть имя)
            return inner_name.split()[0]
            
    # 2. Если незамужняя девушка "Miss. Имя" или женщина без скобок
    # Отрезаем всё, что идет после обращения (Miss., Mrs., Mlle. и т.д.)
    # Структура: Фамилия, Обращение. Имя
    match = re.search(r'(?:Miss\.|Mrs\.|Mlle\.|Mme\.)\s+([A-Za-z]+)', full_name)
    if match:
        return match.group(1)
        
    return None

# Отбираем только строки с женщинами
female_names = df[df['Sex'] == 'female']['Name']

# Применяем функцию извлечения имен и очищаем от пустых значений
extracted_names = female_names.apply(get_female_first_name).dropna()

# Находим самое часто встречающееся имя
ans6 = extracted_names.value_counts().idxmax()
save_answer('6.txt', ans6)
```

```python
# Удалить ответы
import os

# Список файлов для удаления
files_to_delete = [f"{i}.txt" for i in range(1, 7)]

# Удаление файлов
for file in files_to_delete:
    if os.path.exists(file):
        os.remove(file)
        print(f"Файл {file} удалён.")
    else:
        print(f"Файл {file} не найден.")
```
