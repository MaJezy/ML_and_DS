```python
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# 1. Загрузка данных
train_df = pd.read_csv('perceptron-train.csv')
test_df = pd.read_csv('perceptron-test.csv')

# Разделяем на признаки и целевую переменную
X_train = train_df.drop(columns=['Res'])
y_train = train_df['Res']

X_test = test_df.drop(columns=['Res'])
y_test = test_df['Res']

# 2. Обучение персептрона ДО нормализации
# Не забываем зафиксировать random_state=241 по инструкции
clf = Perceptron(random_state=241)
clf.fit(X_train, y_train)

# 3. Подсчет качества (accuracy) на тестовой выборке до нормализации
preds_before = clf.predict(X_test)
acc_before = accuracy_score(y_test, preds_before)

# 4. Нормализация признаков при помощи StandardScaler
scaler = StandardScaler()
# На обучающей выборке считаем параметры и трансформируем
X_train_scaled = scaler.fit_transform(X_train)
# Тестовую выборку только трансформируем по параметрам train!
X_test_scaled = scaler.transform(X_test)

# 5. Обучение персептрона НА ПОСЛЕ нормализации
clf_scaled = Perceptron(random_state=241)
clf_scaled.fit(X_train_scaled, y_train)

# Находим долю правильных ответов на тестовой выборке после нормализации
preds_after = clf_scaled.predict(X_test_scaled)
acc_after = accuracy_score(y_test, preds_after)

# 6. Вычисление разности (После - До)
# Округляем дробную часть до трех знаков, как просит инструкция
diff = acc_after - acc_before
ans_str = f"{diff:.3f}"

# Сохранение ответа в файл без перевода строки на конце
with open('perceptron_answer.txt', 'w', encoding='utf-8') as f:
    f.write(ans_str.strip())

# Вывод результатов в консоль для контроля
print("--- Результаты работы персептрона ---")
print(f"Accuracy до нормализации: {acc_before:.3f}")
print(f"Accuracy после нормализации: {acc_after:.3f}")
print(f"Разность (ответ для Coursera): {ans_str}")
```
